### Natural Filtering

```python
import numpy as np
import numpy.random as rnd
import scipy.stats as stat
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Parameters
T = 24
mu = 5
phi = 0.9
nu = 0.5
sigma = 1
```

```python
# Generating true states
x_true = np.ones(T)
y = np.ones(T)

x_true[0] = rnd.normal(loc = mu, scale = nu/np.sqrt((1-phi**2)))
y[0] = x_true[0] + sigma*rnd.normal()

for t in range(1,T):
    x_true[t] = mu + phi*(x_true[t-1] - mu) + nu*rnd.normal()
    y[t] = x_true[t] + sigma*rnd.normal()
```

```python
# Visualizing data
plt.figure(figsize = (12,6))
plt.plot(x_true, 'x')
plt.plot(y, 'o')
plt.legend(['True states','Observed states'])
plt.grid(True)
plt.xlabel("t")
plt.show()
```

```python
# Allocating memory
N = 100
W = np.zeros((N, T))
X = np.zeros((N, T))
ESS = np.zeros(T)

# First draw
X[:, 0] = rnd.normal(mu, nu/(1-phi**2), N)
W[:, 0] = stat.norm.pdf(y[0], loc = X[:,1], scale = sigma)
W[:, 0] /= (W[:,0].sum())
```

```python
# Main loop
for t in range(1,T):
    X[:, t] = rnd.normal(mu + phi*(X[:,t-1] - mu), nu)
    log_weight = np.log(W[:, t-1]) + np.log(stat.norm.pdf(y[t], loc = X[:, t], scale = sigma))
    W[:, t] = np.exp(log_weight)
    W[:, t] /= W[:, t].sum()
    
    # Computing effective sample size using the approximation from lecture notes
    ESS[t] = 1/((W[:,t]**2).sum())
```

```python
# Just a cute exercise visualizing particles
for i in range(N):
    plt.plot(X[i,:])
    
plt.title('Particles under natural filtering')
plt.xlabel('t')
plt.show()
```

```python
# Let's visualize weights
fig, ax = plt.subplots(12, figsize = (15,30))

for q in range(0, 12):
    ax[q].stem(W[:, 2*q])
    ax[q].set_title('Weights at n={0} under natural filtering '.format(2*q))
    
#plt.show()
```

```python
# Let's visualize ESS
plt.plot(ESS, '-')
plt.grid(True)
#plt.show
```

### General Filter - Resampling

```python
# Let us define the resampling method, using 'stratified sampling', which is said to be more efficient
def resample(weights):
    n = len(weights)
    emp_dist = np.concatenate(([0.], weights.copy().cumsum()))
    u0 = rnd.random()
    j = 0
    index = np.zeros(n, dtype=int)
    for i in range(n):    # Each i is a particle
        u = (u0 + i)/n
        #j = 0
        while u > emp_dist[j]:    # We update j until it crosses the empirical distribution
            j += 1
        index[i] = j-1
    return index
```

```python
# The initial steps are exactly the same as before
X[:,0] = mu + nu/np.sqrt(1-phi**2) * rnd.normal(size=N)
W[:, 0] = stat.norm.pdf(y[0], loc = X[:,0], scale = sigma)
W[:, 0] /= (W[:,0].sum())
ESS[0] = 1/(W[0,:]**2).sum()
```

```python
# Again, the main loop
for t in range(1,T):
    # We start each iteration resampling from previous empirical distribution
    new_shot = resample(W[:,t-1])
    
    # Sampling from updated proposal distribution
    X[:, t] = mu + phi*(X[new_shot, t-1] - mu) + nu*rnd.normal(size=N)
    log_weight = np.log(stat.norm.pdf(y[t], loc = X[new_shot, t], scale = sigma))
    W[:, t] = np.exp(log_weight)
    W[:, t] /= W[:, t].sum()
    ESS[t] = 1/((W[:,t]**2).sum())
```

```python
# Same exercise visualizing particles
for i in range(N):
    plt.plot(X[i,:])
    
plt.title('Particles under stratified resampling')
plt.xlabel('t')
plt.show()
```

```python
# Checking weights
fig, ax = plt.subplots(12, figsize = (15,30))

for k in range(0, 12):
    ax[k].stem(W[:, 2*k])
    ax[k].set_title('Weights at n={0} - stratified sampling '.format(2*k))
    
plt.show()
```
